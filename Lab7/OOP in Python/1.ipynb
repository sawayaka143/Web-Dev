```python
class Vehicle:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year
        self.engine_started = False

    def start_engine(self):
        self.engine_started = True
        return f"The {self.year} {self.make} engine is now running."

    def stop_engine(self):
        self.engine_started = False
        return f"The {self.year} {self.make} engine is now off."

    def __str__(self):
        return f"{self.year} {self.make} {self.model}"
```

```python
class Car(Vehicle):
    def __init__(self, make, model, year, num_doors):
        super().__init__(make, model, year)
        self.num_doors = num_doors

    def open_trunk(self):
        return f"The trunk of the {self.model} is now open."

    def start_engine(self):
        status = super().start_engine()
        return status 

class Truck(Vehicle):
    def __init__(self, make, model, year, payload_capacity):
        super().__init__(make, model, year)
        self.payload_capacity = payload_capacity

    def haul_load(self, weight):
        if weight <= self.payload_capacity:
            return f"Hauling {weight}kg successfully."
        return "Load exceeds payload capacity."

    def start_engine(self):
        status = super().start_engine()
        return status
```

```python
def main():
    my_car = Car("Tesla", "Model 3", 2024, 4)
    my_truck = Truck("Ford", "F-150", 2023, 1000)

    fleet = [my_car, my_truck]

    for vehicle in fleet:
        print(f"Vehicle: {vehicle}")
        print(f"Action: {vehicle.start_engine()}")
        
        if isinstance(vehicle, Car):
            print(f"Feature: {vehicle.open_trunk()}")
        elif isinstance(vehicle, Truck):
            print(f"Feature: {vehicle.haul_load(500)}")
        print("-" * 20)

if __name__ == "__main__":
    main()
```
