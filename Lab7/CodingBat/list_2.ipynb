```python
def count_evens(nums):
    count = 0
    for n in nums:
        if n % 2 == 0:
            count += 1
    return count
```

```python
def sum13(nums):
    if not nums:
        return 0
    total = 0
    skip = False
    for i in range(len(nums)):
        if nums[i] == 13:
            skip = True
            continue
        if skip:
            skip = False
            continue
        total += nums[i]
    return total
```

```python
def big_diff(nums):
    return max(nums) - min(nums)
```

```python
def sum67(nums):
    total = 0
    ignore = False
    for n in nums:
        if n == 6:
            ignore = True
            continue
        if ignore:
            if n == 7:
                ignore = False
            continue
        total += n
    return total
```

```python
def centered_average(nums):
    total = sum(nums) - max(nums) - min(nums)
    return total // (len(nums) - 2)
```

```python
def has22(nums):
    for i in range(len(nums) - 1):
        if nums[i] == 2 and nums[i+1] == 2:
            return True
    return False
```
