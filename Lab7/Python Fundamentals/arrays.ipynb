```python
# Problem A

n = int(input())
l = list(map(int, input().split()))

for i in range(n):
  if i % 2 == 0:
    print(l[i], end=' ')
```

```python
# Problem B

n = int(input())
l = list(map(int, input().split()))

for el in l:
  if el % 2 == 0:
    print(el, end=' ')
```

```python
# Problem C

n = int(input())
l = list(map(int, input().split()))

cnt = 0

for el in l: 
  if el > 0:
    cnt += 1
    
print(cnt)
```

```python
# Problem D

n = int(input())
l = list(map(int, input().split()))

cnt = 0

for i in range(1, n):
  if l[i] > l[i - 1]:
    cnt += 1
    
print(cnt)
```

```python
# Problem E

n = int(input())
l = list(map(int, input().split()))

for i in range(1, n):
    if l[i] == l[i - 1]:
        print("YES")
        break
else:
    print("NO")
```

```python
# Problem F
n = int(input())
l = list(map(int, input().split()))

count = 0

for i in range(1, n - 1):
    if l[i] > l[i - 1] and l[i] > l[i + 1]:
        count += 1

print(count)
```

```python
# Problem G
n = int(input())
l = list(map(int, input().split()))

for i in range(n // 2):
    l[i], l[n - 1 - i] = l[n - 1 - i], l[i]

print(*l)
```
