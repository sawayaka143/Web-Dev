```python
# Problem A

def min4(a, b, c, d):
    if a < b:
        m1 = a
    else:
        m1 = b
    
    if c < d:
        m2 = c
    else:
        m2 = d
    
    if m1 < m2:
        return m1
    else:
        return m2

numbers = list(map(int, input().split()))

result = min4(numbers[0], numbers[1], numbers[2], numbers[3])

print(result)
```

```python
# Problem B

def power(a, n):
    res = 1.0
    for i in range(n):
        res *= a
    return res

input_data = input().split()
a = float(input_data[0])
n = int(input_data[1])

print(power(a, n))
```

```python
# Problem C

def xor(x, y):
    if x != y:
        return True
    else:
        return False

input_data = input().split()
x = bool(int(input_data[0]))
y = bool(int(input_data[1]))

result = xor(x, y)
print(int(result))
```
