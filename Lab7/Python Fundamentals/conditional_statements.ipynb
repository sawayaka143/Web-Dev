```python
# Problem A
a, b = map(int, input().split())
if a > b:
  print(a)
else: 
  print(b)
```

```python
# Problem B
year = int(input())
if (year % 4 == 0) and (year % 100 != 0 or year % 400 == 0):
  print("YES")
else:
  print("NO") 
```

```python
# Problem C
a, b = map(int, input().split())
res = "YES" if (a == 1 and b == 1) or (a != 1 and b != 1) else "NO"
print(res)
```

```python
# Problem D
a = int(input())
if a > 0:
  print(1)
elif a < 0:
  print(-1)
else:
  print(0)
```

```python
# Problem E
a, b = map(int, input().split())
if a > b:
  print(1)
elif b > a:
  print(2)
else:
  print(0)
```
