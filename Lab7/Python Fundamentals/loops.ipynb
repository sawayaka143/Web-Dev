```python
# Problem A
a, b = map(int, input().split())
for i in range(a, b+1):
  if i % 2 == 0:
    print(i, end=' ')
```

```python
# Problem B
a, b, c, d = map(int, input().split())
for i in range(a, b):
  if i % d == c:
    print(i, end=' ')
    
```

```python
# Problem C
a, b = map(int, input().split())

for i in range(1, b + 1):
    square = i * i
    if square < a:
        continue
    if square > b:
        break
    print(square, end=" ")
```

```python
# Problem D
x = int(input())
d = int(input())

count = 0

if x == 0 and d == 0:
    count = 1
else:
    for _ in range(100):
        if x == 0:
            break
        last_digit = x % 10
        if last_digit == d:
            count += 1
        x //= 10
print(count)
```

```python
# Problem E
x = int(input())
total_sum = 0

for i in range(1000):
    if x == 0:
        break
    total_sum += x % 10
    x //= 10

print(total_sum)
```

```python
# Problem A

num = int(input())
i = 1

while i * i <= num:
  print(i*i, end=' ')
  i += 1  
```

```python
# Problem B

n = int(input())
i = 2

while n:
  if n % i == 0:
    print(i)
    break
  i += 1
```

```python
# Problem C

n = int(input())
i = 1

while i <= n:
  print(i, end=' ')
  i *= 2   
```

```python
# Problem D

n = int(input())

while n % 2 == 0:
  n = n // 2
  
if n == 1:
  print("YES")
else:
  print("NO")
```

```python
# Problem E 

n = int(input())
# 7
i = 1
k = 0

while i < n:
  i *= 2
  k += 1
  
print(k)
```
